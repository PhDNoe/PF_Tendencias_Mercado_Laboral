### Normalizamos ambas tablas (Solo los nombres de las columas y los nombres de las provincias)

---

```python
def get_differences(df_2022, df_2023):
    # Obtenemos las columnas de cada dataset
    col_2022 = df_2022.columns.tolist()
    col_2023 = df_2023.columns.tolist()

    print("Cantidad columnas 2022 --> ", len(col_2022))
    print("Cantidad columnas 2023 --> ", len(col_2023))

    # Creamos un set para cada uno
    col_2022_set = set(col_2022)
    col_2023_set = set(col_2023)

    # Realizamos la interseccino
    intersection_set = col_2022_set.intersection(col_2023_set)
    print("Cantidad de columnas en la interseccion --> ", len(list(intersection_set)))

    # Columnas que no estan en ambos conjuntos
    dif_col = col_2022_set.difference(col_2023_set)
    print(dif_col)

get_differences(df_2022, df_2023)
```

```python
def normalize_df(df_2022, df_2023):
    # Renombramos las columnas, quitando los espacios en blanco
    for c in df_2023.columns:    
        df_2023.rename(str.strip, axis='columns',inplace=True)

    for c in df_2022.columns:    
        df_2022.rename(str.strip, axis='columns', inplace=True)

    # Hay un acento de diferencia en ambas columnas
    df_2022.rename(columns={'Si trabajas bajo un esquema híbrido ¿Cuántos días a la semana vas a la oficina?':'Si trabajás bajo un esquema híbrido ¿Cuántos días a la semana vas a la oficina?'}, inplace=True)

normalize_df(df_2022, df_2023)
```

```python
# volvemos a verificar las diferencias
get_differences(df_2022, df_2023)

# Recordar que los dataframes se pasan por referencia y no por valor
```

```python
# Renombramos las provincias para una mejor visualizacion
def rename_provincias(df):

    df['Dónde estás trabajando'] = df['Dónde estás trabajando'].map(lambda x: x.replace('Ciudad Autónoma de Buenos Aires','CABA'))
    df['Dónde estás trabajando'] = df['Dónde estás trabajando'].map(lambda x: x.replace('Provincia de Buenos Aires','BS AS'))
    df['Dónde estás trabajando'] = df['Dónde estás trabajando'].map(lambda x: x.replace('Santiago del Estero','S. del Estero'))


rename_provincias(df_2022)
rename_provincias(df_2023)
```

```python
def get_pais(df, show=True):
    paises = df['Estoy trabajando en'].unique().tolist()
    if show:
        print("-------------------------------------------")
        print("Paises:")
        for p in paises:
            print(" - ",p)
        print("-------------------------------------------")
    return paises
```

#### Distribución de Cantidad de empleos por provincia

---

```python
def show_provincia(df,title):
    fig, ax = plt.subplots()
    df['Dónde estás trabajando'].value_counts().plot(kind='bar', xlabel='Provincia', ylabel='Count', rot=45, ax=ax)
    ax.set(title=title)
    plt.show()

show_provincia(df_2022, "Distribucion de empleos por provincias en el año 2022")
show_provincia(df_2023, "Distribucion de empleos por provincias en el año 2023")
```

---
#### Distribución full time - part-time

---

```python
def show_dedicacion(df,title):
    fig, ax = plt.subplots()
    df['Dedicación'].value_counts().plot(kind='bar', xlabel='Dedicacion', ylabel='Count', rot=45, ax=ax)
    ax.set(title=title)
    plt.show()


show_dedicacion(df_2022,"Cantidad de empleos part-time/full-time año 2022")
show_dedicacion(df_2023,"Cantidad de empleos part-time/full-time año 2023")
```
