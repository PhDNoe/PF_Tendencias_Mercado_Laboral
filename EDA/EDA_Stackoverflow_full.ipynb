```python
def load_data_and_dict(file, year, df_dict):
    df = pd.read_csv(file)
    
    aux = df.columns.tolist()
    aux.sort()
    df_dict[year] = {
        'Columns':aux,
        'n_rows':df.shape[0],
        'n_cols':df.shape[1]    
    }

    print("---------------------------------------------------------------")
    print("Year: ", year)
    print("Cantidad de filas --> ", df_dict[year]['n_rows'])
    print("Cantidad de columnas --> ", df_dict[year]['n_cols'])
    print("---------------------------------------------------------------")
    
    return df, df_dict
```
