```python
# Debemos pasar todo a dolares

currency_dict

# Informacion obtenida de investing.com  --> datos historicos

exchange_dict = {
    'BRL': 5.51, # 1 dolar --> 5.51 reales (promedio en enero 2022),
    'UYU': 44.48,
    'USD':1,
    'ARS': 104, ## Dolar oficial
    'MXN': 20.49,
    'CRC': 636.01,
    'PEN': 3.88,
    'COP': 3990.64,
    'EUR': 0.8836,
    'DKK': 6.5753,
    'CAD': 1.2630,
    'XAF': 586.786,
    'GBP':  0.7377   
    
    # Hay que seguir buscando, uno por uno
}

code_list_prev = ['BRL', 'UYU', 'USD', 'ARS', 'MXN', 'CRC', 'PEN', 'COP']
country_list_prev = ['Brazil', 'Uruguay', 'United States of America',\
                    'Argentina', "Mexico","Costa Rica", "Peru", "Colombia"] # previsorias
```

```python
import matplotlib.pyplot as plt
import numpy as np

def convert_salary(row, conversion_rates, target_currency):
    currency = row['Currency']
    salary = row['CompTotal']
    
    if currency == target_currency:
        return salary
    
    if currency in conversion_rates:
        rate = conversion_rates[currency]
        return round(salary / rate,2)
    
    return np.nan
```

```python
# Creamos un df con los paises que he llegado a convertir

mask1 = df_latin['Country'].isin(country_list_prev)
df_short = df_latin.loc[mask1]
```

```python
df_short.Country.unique()
```

```python
df_short.isna().sum().sort_values()
```

```python
# Aplico la funcion
target_currency = "USD"
df_short["converted_salary"] = df_short.apply(convert_salary, conversion_rates=exchange_dict, target_currency=target_currency, axis=1)
df_short.head()
```

```python
df_short.loc[df_short['converted_salary'].isna(),["Country","Currency","CompTotal", "CompFreq"]]
```

```python

df_sueldos = df_short[['Currency',	'converted_salary',	'CompFreq']]
df_sueldos2 = df_short[['Country',	'converted_salary',	'CompFreq']]

df_anual = df_sueldos.loc[df_sueldos['CompFreq']=='Yearly']
df_anual2 = df_sueldos2.loc[df_sueldos2['CompFreq']=='Yearly']

df_anual['Currency'] = df_anual['Currency'].map(currency_dict)
# df_anual['Currency'] = df_anual['Currency'].map(currency_dict)

por_moneda = df_anual.groupby('Currency')['converted_salary'].median().sort_values(ascending=False)[:10]
por_pais = df_anual2.groupby('Country')['converted_salary'].median().sort_values(ascending=False)[:10]
por_pais_cuenta = df_anual2.groupby('Country')['converted_salary'].count().sort_values(ascending=False)[:10]

fig, ax = plt.subplots()
ax.bar(por_moneda.index, por_moneda.values)
plt.xticks(rotation=90)
ax.set(title="Sueldo promedio anual por Moneda original (en dolares) (Mediana)")
ax.set(xlabel="Pais",ylabel="Sueldo anual [USD]")
# Ajustar los parámetros de las etiquetas del eje x
ax.tick_params(axis='x', which='both', labelsize=10, pad=5)


fig, ax = plt.subplots()
ax.bar(por_pais.index, por_pais.values)
plt.xticks(rotation=90)
ax.set(title="Sueldo promedio anual por paises (en dolares) (Mediana)")
ax.set(xlabel="Pais",ylabel="Sueldo anual [USD]")
# Ajustar los parámetros de las etiquetas del eje x
ax.tick_params(axis='x', which='both', labelsize=10, pad=5)

fig, ax = plt.subplots()
ax.bar(por_pais_cuenta.index, por_pais_cuenta.values)
plt.xticks(rotation=90)
ax.set(title="Cantidad de empleos por paises")
ax.set(xlabel="Pais",ylabel="Cantidad")
# Ajustar los parámetros de las etiquetas del eje x
ax.tick_params(axis='x', which='both', labelsize=10, pad=5)
```

```python
# Agrupar por país y contar la cantidad de empleos
por_pais_cuenta = df_anual2.groupby('Country')['converted_salary'].count().sort_values(ascending=False)


# Definir la cantidad mínima de empleos por país para considerarlo
min_empleos_por_pais = 15

# Filtrar para quedarse solo con los países con la cantidad mínima de empleos
por_pais_filtrado = por_pais_cuenta[por_pais_cuenta >= min_empleos_por_pais]

# Hacer la mediana de los salarios convertidos para los países filtrados y ordenarlos por valor descendente
por_pais = df_anual2[df_anual2['Country'].isin(por_pais_filtrado.index)].groupby('Country')['converted_salary'].median().sort_values(ascending=False)[:10]


fig, ax = plt.subplots()
ax.bar(por_pais.index, por_pais.values)
plt.xticks(rotation=90)
ax.set(title="Sueldo promedio anual por paises (en dolares) (Mediana) (paises con mas de 15 ofertas)")
ax.set(xlabel="Pais",ylabel="Sueldo anual [USD]")
# Ajustar los parámetros de las etiquetas del eje x
ax.tick_params(axis='x', which='both', labelsize=10, pad=5)
```
